## **1. Social Media – Post Engagement Analyzer**

```python
def post_engagement(likes):
    total_likes = 0

    for like in likes:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_engagement([200, 300, 250, 300])
```

## **2. Healthcare – Medicine Stock Alert System**

```python
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

check_stock(6)
```

## **3. Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_check(rainfall, required):
    total = 0

    for r in rainfall:
        total += r

    average = total / len(rainfall)

    if average >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_check([70, 75, 80, 63], 70)
```

## **4. Social Media – Duplicate Account Detection**

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

check_duplicates(["alice", "bob", "alice"])
```

## **5. Healthcare – Appointment Eligibility Checker**

```python
def appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

## **6. Agriculture – Premium Crop Price Filter**

```python
def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

## **7. System Monitoring – Application Health Checker**

```python
def app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

app_health(7)
```

## **8. Banking – Daily Transaction Limit Checker**

```python
def transaction_check(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

## **9. E-Learning – Student Attendance Eligibility System**

```python
def attendance_check(attendance):
    total = 0

    for day in attendance:
        total += day

    percentage = (total / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_check([1, 1, 0, 1, 1])
```

## **10. Smart Electricity Bill Analyzer**

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", usage)

electricity_bill(250)
```
