## **1. Mobile Recharge Validation System**

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Invalid: Amount must be at least ₹50")
            continue

        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid plan. Try again.")

validate_recharge()
```

## **2. Inventory Reorder Alert System**

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

n = int(input("Enter number of products: "))
products = {}

for i in range(n):
    name = input("Enter product name: ")
    stock = int(input("Enter stock quantity: "))
    products[name] = stock

check_inventory(products)
```

## **3. Student Result Processing System**

```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = list(map(int, input("Enter marks: ").split()))

result = process_result(marks)
print("Result:", result)
```

## **4. Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)
    if peak == "yes":
        fare += fare * 0.25
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ").lower()

    fare = calculate_fare(distance, peak)
    print("Total Fare:", fare)

    retry = input("Calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

## **5. Employee Attendance Eligibility Checker**

```python
def check_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = input("Enter attendance: ").split()

result = check_eligibility(attendance)
print("Attendance Status:", result)
```

## **6. Password Strength Checker**

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter password: ")
print(check_password(password))
```
