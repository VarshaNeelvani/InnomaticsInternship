## **1. Unique Words in a Sentence**

```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## **2. Highest Salary from Employee Data**

```python
n = int(input("Enter number of employees: "))

employees = {}
for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

## **3. Find Maximum and Minimum Values**

```python
numbers = list(map(int, input("Enter numbers: ").split()))

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## **4. Count Products Above a Price Threshold**

```python
prices = list(map(int, input("Enter product prices: ").split()))

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## **5. Calculate Attendance Percentage**

```python
attendance = input("Enter attendance: ").split()

present_count = attendance.count("P")
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

## **6. Remove Duplicate Phone Numbers**

```python
phone_numbers = list(map(int, input("Enter phone numbers: ").split()))

unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

## **7. Count Character Frequency**

```python
text = input("Enter a string: ")

frequency = {}
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

## **8. Convert List to Tuple**

```python
numbers = list(map(int, input("Enter numbers: ").split()))

t = tuple(numbers)
print("Tuple:", t)
```

## **9. Check if a Key Exists in a Dictionary**

```python
n = int(input("Enter number of employees: "))

employees = {}
for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

key_to_check = input("Enter employee name to search: ")

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

## **10. Calculate Average Marks**

```python
marks = list(map(int, input("Enter marks: ").split()))

average = sum(marks) / len(marks)
print("Average Marks:", average)
```
