## **1. Employee Performance Bonus Eligibility**

```python
n = int(input("Enter number of employees: "))

employees = {}
for i in range(n):
    name = input("Enter employee name: ")
    score = int(input("Enter performance score: "))
    employees[name] = score

highest_score = max(employees.values())

top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

## **2. Search Query Keyword Analysis**

```python
import string

query = input("Enter search query: ").lower()

# remove punctuation
for p in string.punctuation:
    query = query.replace(p, "")

words = query.split()

frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# show only words with frequency > 1
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

## **3. Sensor Data Validation**

```python
sensor_readings = list(map(int, input("Enter sensor readings: ").split()))

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## **4. Email Domain Usage Analysis**

```python
n = int(input("Enter number of emails: "))

emails = []
for i in range(n):
    emails.append(input("Enter email: "))

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## **5. Sales Spike Detection**

```python
sales = list(map(int, input("Enter daily sales: ").split()))

average_sales = sum(sales) / len(sales)
threshold = round(average_sales * 1.30)

for i in range(len(sales)):
    if sales[i] >= threshold:
        print(f"Day {i+1}: {sales[i]}")
```

## **6. Duplicate User ID Detection**

```python
user_ids = input("Enter user IDs: ").split()

frequency = {}

for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
