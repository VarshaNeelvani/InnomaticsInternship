## Simple Chain

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

chain = prompt | llm

response = chain.invoke({"topic": "AI Agents"})
print("Chain Output:", response.content)
```

##Memory Example

```python
class SimpleMemory:
    def __init__(self):
        self.history = []

    def save_context(self, user_input, output):
        self.history.append(f"Human: {user_input['input']}")
        self.history.append(f"AI: {output['output']}")

    def load_memory(self):
        return "\n".join(self.history)


memory = SimpleMemory()

memory.save_context({"input": "Hi"}, {"output": "Hello!"})
memory.save_context({"input": "How are you?"}, {"output": "I'm good!"})

print(memory.load_memory())
```

```python
def calculator(expression):
    try:
        return str(eval(expression))
    except:
        return "Error"

query = "What is 15 * 6?"

print("User Query:", query)

if "*" in query or "+" in query or "-" in query or "/" in query:
    expression = "15 * 6"  # extracted manually for demo
    result = calculator(expression)
    print("Agent used Calculator Tool")
    print("Final Answer:", result)
else:
    response = llm.invoke(query)
    print("LLM Answer:", response.content)
```
