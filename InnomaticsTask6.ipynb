## **1. Smart Parking Lot Management System**

```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

## **2. Online Food Delivery Time Estimator**

```python
def delivery_time(distance, traffic, weather):
    # Base time: 5 minutes per km
    eta = distance * 5

    # Traffic adjustment
    if traffic == "High":
        eta += 10
    elif traffic == "Medium":
        eta += 5

    # Weather adjustment
    if weather == "Rainy":
        eta += 5
    elif weather == "Storm":
        eta += 10

    print("Estimated Delivery Time:", eta, "minutes")

delivery_time(8, "High", "Rainy")
```

## **3. Movie Theatre Seat Occupancy**

```python
def seat_analyzer(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Show Status:", status)

seat_analyzer(200, [1]*150)
```

## **4. Cloud Server Load Classification System**

```python
def server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", average, "%")

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

## **5. Smart Classroom Resource Usage Monitor**

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

## **6. Online Event Registration Capacity Controller**

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100, 105)
```
