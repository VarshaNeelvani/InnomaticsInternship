## 2. **Pass / Fail Analyzer**

```python
marks = [45, 78, 90, 33, 60]

passed_students = 0
failed_students = 0

for i in marks:
  if i >= 50:
    passed_students += 1
  else:
    failed_students += 1

print("Number of Passed Students: ", passed_students)
print("Number of Failed Students: ", failed_students)
```

## 3. **Simple Data Cleaner**

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print(cleaned_names)
```

## 4. **Message Length Analyzer**

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print("Length:", len(msg))
    if len(msg) > 10:
        print("Message longer than 10 characters")
```

##5. **Error Message Detector**

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
